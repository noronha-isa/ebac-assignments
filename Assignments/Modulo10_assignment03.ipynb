**2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?**

````python
# Perfis médios no tempo

fig, axs = plt.subplots(2,1, figsize=(15,10))

ax0 = sns.pointplot(x='data_ref', y='renda', hue='educacao', data=df, dodge=True, ci=95, ax=axs[0])
ax1 = sns.pointplot(x='data_ref', y='renda', hue='estado_civil', data=df, dodge=True, ci=95, ax=axs[1])
````

````python
fig, axs = plt.subplots(2,1, figsize=(15,10))

ax2 = sns.pointplot(x='data_ref', y='renda', hue='sexo', data=df, dodge=True, ci=95, ax=axs[0])
ax3 = sns.pointplot(x='data_ref', y='renda', hue='mau', data=df, dodge=True, ci=95, ax=axs[1])
````

````python
fig, axs = plt.subplots(2,1, figsize=(15,10))

ax4 = sns.pointplot(x='data_ref', y='renda', hue='tipo_residencia', data=df, dodge=True, ci=95, ax=axs[0])
ax5 = sns.pointplot(x='data_ref', y='renda', hue='tipo_renda', data=df, dodge=True, ci=95, ax=axs[1])
````
